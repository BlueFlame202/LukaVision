```python
from ultralytics import YOLO
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
model = YOLO("yolov8m.pt")
```

```python
image = "../../data/roboflow/Basketball Players.v22-raw-images-scoreboardclassesonly-nonulls.yolov8/train/images/youtube-5_jpg.rf.fe97e3b43771ec1cd42472efe9b3ecfc.jpg"
```

```python
results = model.predict(image)
results
```

```python
result = results[0]
result
```

```python
len(result.boxes)
```

```python
box = result.boxes[0] 
box

"""
xyxy - the coordinates of the box as an array [x1,y1,x2,y2]
cls - the ID of object type
conf - the confidence level of the model about this object. If it's very low, like < 0.5, then you can just ignore the box.
"""
```

```python
print("Object type:", box.cls)
print("Coordinates:", box.xyxy)
print("Probability:", box.conf)
```

```python
for box in result.boxes:
  class_id = result.names[box.cls[0].item()]
  cords = box.xyxy[0].tolist()
  cords = [round(x) for x in cords]
  conf = round(box.conf[0].item(), 2)
  print("Object type:", class_id)
  print("Coordinates:", cords)
  print("Probability:", conf)
  print("---")
```

```python
vid_path = r"../../data/kawhivideovision.mp4"
results = model.track(source = vid_path, show = True, tracker = "bytetrack.yaml")
```
